```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
breed = ['Abyssinian','American Bulldog', 'American Pitbull Terrier', 'Basset Hound', 'Beagle', 'Bengal', 'Birman', 'Bombay',
           'Boxer', 'British Shorthair', 'Chihuahua', 'Egyptian Mau' , 'English Cocker Spaniel' , 'Englers Setter' , 'German Shorthair',
           'Great Pyrenees', 'Havanese' , 'Japanese Chin' , 'Keeshond', 'Leonberger', 'Maine Coon', 'Minature Pinscher', 'Newfoundland',
           'Persian', 'Pomeranian', 'Pug', 'Ragdoll', 'Russian Blue', 'Saint Bernard', 'Samoyed', 'Scottish Terrier', 'Shiba Inu',
           'Siamese', 'Sphynx', 'Staffordshire Bull Terrier', 'Wheaton Terrier', 'Yorkshire Terrier']
#print(len(breed))
counts = [203, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200, 200,
          200, 200, 200, 200, 200, 200, 200, 200, 200, 199, 200, 200, 200, 191, 200, 200]
#print(len(counts))
species = ['cat', 'dog', 'dog', 'dog', 'dog', 'cat', 'cat', 'cat', 'dog', 'cat', 'dog','cat', 'dog', 'dog' ,'dog', 'dog',
           'dog', 'dog', 'dog', 'dog', 'cat', 'dog', 'dog', 'cat', 'dog', 'dog', 'cat', 'cat', 'dog', 'dog', 'dog', 'dog',
           'cat', 'cat', 'dog', 'dog', 'dog']
#print(len(species))

df = pd.DataFrame({'Species': species, 'Breed': breed, 'Count': counts})
df.head(37)
```

```python
grouped_df = df[['Species', 'Count']].groupby('Species').sum()

# make chart
plt.pie(grouped_df['Count'], labels=grouped_df.index, colors=sns.color_palette('dark'), autopct='%.0f%%')
# now adjust chart
fig = plt.gcf()
for text in fig.gca().texts:
    text.set_color('white')
# Remove background
fig.set_facecolor("None")
```

```python
count_of_categories_df = df.groupby('Species')['Breed'].count().reset_index()
count_of_categories_df.columns = ['Species', 'Total_Breeds']
count_of_categories_df.head(2)
```
