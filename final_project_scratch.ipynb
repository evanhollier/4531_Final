```python
# Make function to reindex the photos
def reindex_photos(dir, category):
    current_dir = os.path.join(dir, category)
    photo_list = os.listdir(current_dir)
    for i in range(len(photo_list)):
        os.rename(os.path.join(current_dir, photo_list[i]), os.path.join(current_dir, f'{category}_{i}.jpg'))

for i in categories:
    reindex_photos(r'c:\Users\Paul\Documents\CS4531\images_sorted', i)
```
